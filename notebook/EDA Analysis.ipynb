```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

```

```python
cal_df=pd.read_csv('./../notebook/data/calories.csv')
exer_df=pd.read_csv('./../notebook/data/exercise.csv')
```

```python
cal_df.head()
```

```python
exer_df.head()
```

```python
df=pd.merge(exer_df,cal_df,on='User_ID',how='inner')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop('User_ID', axis=1, inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Gender',data=df,palette=['blue','pink'])
plt.xlabel('Gender Distribution')
plt.ylabel('Count')
plt.title('Distribution of gender')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Height'], bins=10, color='skyblue', kde=True)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Distribution of Height')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=10, color='red', kde=True)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Distribution of Height')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Weight'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Weight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Heart_Rate'], bins=10, color='darkblue', kde=True)
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Heart rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot to Identify Outliers')
plt.show()
```
